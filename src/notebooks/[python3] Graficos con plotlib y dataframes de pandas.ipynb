```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
matplotlib.rcParams['agg.path.chunksize'] = 10000

data_train = pd.read_csv("data/train.csv", sep=',',encoding='latin1', parse_dates=['Time'], dayfirst=True, index_col='Id')
data_test = pd.read_csv("data/test.csv")

data_train.head()
```

```python
prediction_0 = data_train[data_train['Prediction']==0]['Prediction'].value_counts()
prediction_1 = data_train[data_train['Prediction']==1]['Prediction'].value_counts()
prediction_2 = data_train[data_train['Prediction']==2]['Prediction'].value_counts()
prediction_3 = data_train[data_train['Prediction']==3]['Prediction'].value_counts()
prediction_4 = data_train[data_train['Prediction']==4]['Prediction'].value_counts()
prediction_5 = data_train[data_train['Prediction']==5]['Prediction'].value_counts()

df = pd.DataFrame([prediction_0,prediction_1,prediction_2,prediction_3,prediction_4,prediction_5])
df.index = ['0','1','2','3','4','5']
df.plot(kind='bar',stacked=True, figsize=(7,5))
```

```python
plt.figure(figsize=(18,8))

ax = plt.subplot(121)
ax.scatter(data_train[data_train['Prediction']==1]['Prediction'],data_train[data_train['Prediction']==1]['HelpfulnessNumerator'],c='green',s=40)
ax.scatter(data_train[data_train['Prediction']==2]['Prediction'],data_train[data_train['Prediction']==2]['HelpfulnessNumerator'],c='red',s=40)
ax.scatter(data_train[data_train['Prediction']==3]['Prediction'],data_train[data_train['Prediction']==3]['HelpfulnessNumerator'],c='orange',s=40)
ax.scatter(data_train[data_train['Prediction']==4]['Prediction'],data_train[data_train['Prediction']==4]['HelpfulnessNumerator'],c='brown',s=40)
ax.scatter(data_train[data_train['Prediction']==5]['Prediction'],data_train[data_train['Prediction']==5]['HelpfulnessNumerator'],c='yellow',s=40)
ax.set_xlabel('Prediccion')
ax.set_ylabel('HelpfulnessNumerator')
ax.legend(('1','2','3','4','5'),scatterpoints=1,loc='upper right',fontsize=15,)

ax1 = plt.subplot(122)
ax1.scatter(data_train[data_train['Prediction']==1]['Prediction'],data_train[data_train['Prediction']==1]['HelpfulnessDenominator'],c='green',s=40)
ax1.scatter(data_train[data_train['Prediction']==2]['Prediction'],data_train[data_train['Prediction']==2]['HelpfulnessDenominator'],c='red',s=40)
ax1.scatter(data_train[data_train['Prediction']==3]['Prediction'],data_train[data_train['Prediction']==3]['HelpfulnessDenominator'],c='orange',s=40)
ax1.scatter(data_train[data_train['Prediction']==4]['Prediction'],data_train[data_train['Prediction']==4]['HelpfulnessDenominator'],c='brown',s=40)
ax1.scatter(data_train[data_train['Prediction']==5]['Prediction'],data_train[data_train['Prediction']==5]['HelpfulnessDenominator'],c='yellow',s=40)
ax1.set_xlabel('Prediccion')
ax1.set_ylabel('HelpfulnessDenominator')
ax1.legend(('1','2','3','4','5'),scatterpoints=1,loc='upper right',fontsize=15,)
```
